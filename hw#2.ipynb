```
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import learning_curve

%matplotlib inline
```

```
lr_model = LinearRegression( fit_intercept= True, normalize=True)
```

```
def compute_MSE(h, y):
    return  np.mean((y - h) ** 2) / 2
MSE_TEST = []
MSE_TRAIN = []
score_test  = []
score_train = []
```

```
#daraje 1
x = np.random.rand(200, 1) * 20
y = 2*x**4 - x**3 + 3*x**2 - 5*x + 4
y = y + np.random.normal(scale=3, size=y.shape)
y.reshape(-1,1)
split = int(x.shape[0] * 0.8)
x_train = x[:split]
y_train = y[:split]
x_test = x[split:]
y_test = y[split:]
lr_model.fit(x_train,y_train)
#lr_model.score(x_train, y_train)
h_test = lr_model.predict(x_test)
h_train = lr_model.predict(x_train)
x_line = np.arange(0,20,0.1).reshape(-1,1)
y_line = lr_model.predict(x_line)
plt.plot(x_line[:,0], y_line, 'r--')
plt.plot(x_train,  y_train, 'bo')
plt.plot(x_test,  y_test, 'co')
plt.xlabel("x")
plt.ylabel('y')
plt.title("power 1")
MSE_TEST.append(compute_MSE(h_test, y_test))
MSE_TRAIN.append(compute_MSE(h_train, y_train))
score_test.append(lr_model.score(x_test, y_test))
score_train.append(lr_model.score(x_train, y_train))
```

```
#learning cuve
train_sizes, test_score, train_score = learning_curve(lr_model, x_train[0:200, 1:1+1], y_train, cv=5)
plt.plot(train_sizes, test_score.mean(axis = 1), label = 'validation')
plt.plot(train_sizes, train_score.mean(axis = 1), label = 'train')
plt.legend()
plt.xlabel("training set size")
plt.ylabel("score")
```
